# LEADERBORD

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
classes = ["TeamName","Score"]
art = pd.read_csv("leaderboard.csv")[classes]
art
```

```python
art = art.sort_values("Score",ascending=True)
x = []
y = []
i = 1
for e in art["Score"]:
    x.append(i)
    i += 1
    y.append(e)
```

```python
#lt.style.use("grayscale")
fig, ax = plt.subplots(figsize=(8,8),dpi=300)
fig.patch.set_facecolor('white')
plt.ylim((0,1))
plt.xlim((-100,15000))
plt.plot(x,y)

x2 = np.arange(0,15000,0.1)
y21 = 0.231 * np.ones(150000)
y22 = 0.15297 * np.ones(150000)
plt.plot(x2,y21, "--", label="18. August")
plt.plot(x2,y22, "--", label="21. August")
plt.legend()
```
