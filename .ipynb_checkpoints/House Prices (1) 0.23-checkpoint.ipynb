## 3. Simple Prediction
Given only *GarageCars* and *OverallQual* the predicition for the resulting house price is dependend on each fit weighted given the correlation of each feature, so

$$\frac{cor_{GC}}{cor_{GC}+cor_{OQ}+cor_{GLA}} \cdot fit_{GC}(x_1) + \frac{cor_{OQ}}{cor_{GC}+cor_{OQ}+cor_{GLA}} \cdot fit_{OQ}(x_2)  + \frac{cor_{GLA}}{cor_{GC}+cor_{OQ}+cor_{GLA}} \cdot fit_{GLA}(x_3) = Price,$$

where 

$$ cor_{GC} = 0.640409197258349 $$
$$ cor_{OQ} = 0.7909816005838047 $$
$$ cor_{GLA} = 0.7086244776126511. $$

Only three of the features are considered as this gave they best result.

```python
import math

# root mean squared logarithmic error
def RMSLE(a,p):
    return abs(math.log(a+1) - math.log(p+1))

# prediction function
def pred(x1,x2,x3):
    # given data of the house
    cars = x1  
    rating = x2
    area = x3
    # corrolations
    wgc = 0.640409197258349
    woq = 0.7909816005838047
    wgla = 0.7086244776126511
    # weights
    total = (wgc+woq+wgla)
    w1 = woq/total
    w2 = wgc/total
    w3 = wgla/total
    # predicted price with given data
    p = w1 * func1(rating, *popt1) + w2 * func2(cars, *popt2) + w3 * func3(area, *popt3)
    return p

# show predicition etc.
p = pred(7,2,1710)
a = 208500
print("Predicted price: " + str(p))
print("Actual price: " + str(a))
print("Score: " + str(RMSLE(a,p)))
```

## 4. Prediction Testing

### Training

```python
import math
import time

ACC = 0
for house in range(len(art["Id"])):
    x1 = art["GarageCars"][house]
    x2 = art["OverallQual"][house]
    x3 = art["GrLivArea"][house]
    a = art["SalePrice"][house]
    p = pred(x1,x2,x3)
    acc = RMSLE(a,p)
    ACC += acc

    #print("Accuracy: " + str(acc))

print("Total Training (Mean) Score: " +str(ACC/(len(art["Id"]))))
```

### Testing

```python
classes = ["Id","OverallQual","GarageCars","TotRmsAbvGrd","GrLivArea","YearBuilt"]
art = pd.read_csv("test.csv")[classes]

file = open('submission.csv', 'w')
file.write("Id,SalePrice")
file.close() 

for house in range(len(art["Id"])):
    x1 = art["GarageCars"][house]
    x2 = art["OverallQual"][house]
    x3 = art["GrLivArea"][house]
    p = pred(x1,x2,x3)
    if math.isnan(p):
        print("NaN-Error found and corrected!")
        p = 100000.00
    # write submission csv
    file = open('submission.csv', 'a+')
    file.write("\n" + str(art["Id"][house]) + "," + str(p))
    file.close()

classes = ["Id","SalePrice"]
art = pd.read_csv("submission.csv")[classes]
art
```
