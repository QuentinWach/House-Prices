```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from sklearn.model_selection import train_test_split
from scipy import stats
from scipy.stats import norm, skew
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

## Import Data

1. Import selected data features
2. Normalize data
3. Splitting the training data into training and validation data.

```python
# import
features = ["OverallQual","GarageCars","TotRmsAbvGrd","GrLivArea",
            "YearBuilt","YearRemodAdd","TotalBsmtSF" ,"1stFlrSF", "FullBath", "SalePrice"]
data = pd.read_csv("train.csv")[features]
data.head()
```

```python
# do we have missing data in this set?
b = (data.isnull().sum() / len(data)) * 100
b
```

## Label Normalization
The destribution of the label data, that which we want to generate later might be both biased, meaning its shifted and not in the center of a normal distribution, and also extremely complex having a high degree of variation, meaning that all the points are spread out very far.

![png](./bias-variance.png)

```python
# Density Distribution
fig, ax = plt.subplots(figsize=(12,4),dpi=300)
fig.patch.set_facecolor('white')
fig = sns.distplot(data["SalePrice"], fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(data["SalePrice"])
plt.legend(['$\mu=$ {:.2f}, $\sigma=$ {:.2f}'.format(mu, sigma)],loc='best')

plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig2 = plt.figure()
fig2, ax2 = plt.subplots(figsize=(12,4),dpi=300)
fig2.patch.set_facecolor('white')
ax2 = stats.probplot(data["SalePrice"], plot=plt)
```

```python
#We use the numpy fuction log1p which  applies log(1+x) to all elements of the column
print(data["SalePrice"][:5])
data["SalePrice"] = np.log1p(data["SalePrice"])
print(data["SalePrice"][:5])

#Check the new distribution 
fig, ax = plt.subplots(figsize=(12,4),dpi=300)
fig.patch.set_facecolor('white')
sns.distplot(data["SalePrice"] , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(data["SalePrice"])

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
fig, ax = plt.subplots(figsize=(12,4),dpi=300)
fig.patch.set_facecolor('white')
ax = stats.probplot(data["SalePrice"], plot=plt)
plt.show()

```

## Splitting

As we don't have labels for our test dataset with which we will create the constribution to kaggle, we might overfit our data! In order to have just another way to prevent this we can use our training data and split it! This way we can train and test how general our model is.

```python
# splitting
train_data, val_data, train_label, val_label = train_test_split(data[features[:9]], 
                                                                     data["SalePrice"], 
                                                                     test_size=0.5, random_state=42)

print(len(train_data))
print(len(val_data))
train_data
```

## Model

We train our model and make our predictions. We evaluate it using the RMSE metric of logarithmic prices bla bla bla ...

$$ \operatorname{RMSE}=\sqrt{\frac{\sum_{i=1}^{n}\left(P_{i}-O_{i}\right)^{2}}{n}}$$

```python
# Model testing
def trainAtest(max_leaf_nodes, train_data, val_data, train_label, val_label):
    model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    model.fit(train_data, train_label)
    val_pred = np.exp(model.predict(val_data)) - 1
    val_label = np.exp(val_label) - 1
    mae = np.sqrt(metrics.mean_squared_error(np.log1p(val_label), np.log1p(val_pred)))
    return(mae)

# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5,80,150,500]:
    my_mae = trainAtest(max_leaf_nodes, train_data, val_data, train_label, val_label)
    print(max_leaf_nodes, my_mae)
```

```python
# Final model with optimized leafes

### TRAINING
#=================================================================================================================
# Define model. Specify a number for random_state to ensure same results each run
model = RandomForestRegressor(max_leaf_nodes=70, random_state=1)
model.fit(train_data, train_label)

# Fit model
model.fit(train_data, train_label)
train_pred = model.predict(train_data)

# PREDICTION AND EVALUTAION
#=================================================================================================================
# Predict validation label
val_pred = np.exp(model.predict(val_data)) - 1
val_label = np.exp(val_label) - 1
print(val_pred[:5])
print(val_label[:5])

# Evaluate training and validation score
print("TRAIN SCORE: " + str(np.sqrt(metrics.mean_squared_error(np.log1p(train_label), np.log1p(train_pred)))))
print("VAL SCORE: " + str(np.sqrt(metrics.mean_squared_error(np.log1p(val_label), np.log1p(val_pred)))))
```

## Prediction

```python
### PREDICTING ###
# import test data
features = ["OverallQual","GarageCars","TotRmsAbvGrd","GrLivArea",
            "YearBuilt","YearRemodAdd","TotalBsmtSF" ,"1stFlrSF", "FullBath"]
test_data = pd.read_csv("test.csv")[features]

# do we have missing data in this set?
(test_data.isnull().sum() / len(test_data)) * 100

# # IMPUTING MISSING VALUES
def fill_missing_values(df):
    ''' This function imputes missing values with median for numeric columns 
        and most frequent value for categorical columns'''
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    for column in list(missing.index):
        if df[column].dtype == 'object':
            df[column].fillna(df[column].value_counts().index[0], inplace=True)
        elif df[column].dtype == 'int64' or 'float64' or 'int16' or 'float16':
            df[column].fillna(df[column].median(), inplace=True)
            
fill_missing_values(test_data)
test_data.isnull().sum().max()
```

```python
# predict and renorm with imported test data
test_pred = np.exp(model.predict(test_data)) - 1
test_pred[:5]
```
